```python
import numpy as np
import pandas as pd
```

```python
rot_path= "rotation_matrices/"
error_path="errors/"
names = ["andreadm2.ply", "iacopo2.ply", "subject002.ply", "subject003.ply", "subject004.ply", "subject005.ply", "subject006.ply", "subject007.ply", "subject008.ply", "subject010.ply","subject011.ply", "subject012.ply", "subject013.ply", "subject014.ply","subject015.ply", "subject016.ply", "subject017.ply", "subject018.ply", "subject019.ply", "subject020.ply"]
```

```python
filenames = []
evs = np.zeros(361)
every=[]

for i in names:
    for j in names:
        filename=(error_path+i+"_"+j+".txt")
        file = open(filename, "r")
        count=1
        errs=0
        iters =0
        category =0
        compares=(i+" - "+ j)
        
        
        if(j==i):
            equal = 1
        else:
            equal =0  
            
            
        for line in file:
            if(count==1):
                tmp = float(line)
                iters=int(tmp)
            if(count==2):
                errs = float(line)
            if(count==3):
                tmp2 = float(line)
                category = int(tmp2)
                #category=float(line)
            count=count +1
        filenames.append(filename)
        tmp=np.array([iters, errs, category, equal, compares])
        every.append(tmp)
        
        
complete = np.asarray(every)
```

```python
data = pd.DataFrame({'Iterations': complete[:, 0], 'Error': complete[:, 1], 'Category': complete[:,2], 'Same_face': complete[:,3], 'Compares': complete[:,4]})

data = data.astype({"Iterations": int, "Error": float, 'Category': int, 'Same_face': int, 'Compares': str})

#df2 = data.groupby('Same_face')
#df2['Category']


data.groupby('Category').count()

```

```python
#Rotation matrix to numpy arrays
test = []
tmp4=[]
tmp5=[]
for i in names:
    for j in names:    
        guess_name =i+"_"+j+"_guess.txt"
        init_name=i+"_"+j+"_initial.txt"
        guess_file = open(rot_path+guess_name, "r")
        init_file= open(rot_path+init_name, "r")
        counter = 0
        test=[]
        for line in guess_file:
            tmp = line.rstrip()
            row = tmp.split(",")
            r =  np.array(row, dtype = float)
            counter=counter+1
            test.append(r)
        guess_matrix=np.array(test)
        #print(guess_matrix)
        tmp4.append(guess_matrix)
        test=[]
        for line in init_file:
            tmp = line.rstrip()
            row = tmp.split(",")
            r =  np.array(row, dtype = float)
            counter=counter+1
            test.append(r)
        init_matrix=np.array(test)
        tmp5.append(init_matrix)        
        

initials = np.asarray(tmp5)
guesses =np.asarray(tmp4)
```

```python
diffs = []
for i in range(len(initials)):
    tmp = np.abs(initials[i]-guesses[i])
    tmp2 = tmp.sum()
    diffs.append(tmp2)

#print(diffs)
for n in diffs:
    print (n)
```

```python
data['Matrix difference'] = diffs
```
